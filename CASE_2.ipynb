```python
import pandas as pd
import seaborn as sbs
```

```python
vendas = pd.read_excel("varejo.xlsx")
vendas.head()
```

```python
vendas.info()
```

```python
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```python
# os canais de vendas APP e aplicativo são os mesmo, por isso vamos transforma-los em um só
vendas["idcanalvenda"] = vendas["idcanalvenda"].str.replace("APP","Aplicativo")
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```python
# da mesma forma vamos verificar para os outros parametros
vendas.groupby("bandeira").idcompra.nunique()
```

```python
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
vendas.groupby("estado").idcompra.nunique()
```

```python
# como uma boa prática, vamos substituit os espaços entre os nomes das catergorias por underscore
vendas["Nome_Departamento"] = vendas["Nome_Departamento"].str.replace(" ","_")
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
# tratando os dados nulos, temos duas opções: substitui-los ou exclui-los. como uma das premissas
# é considerar os estados nulos com MS, vamos substitui-los. para apagar temos a função .dropna()

vendas_sem_nulos = vendas.dropna()
vendas_sem_nulos.info()
```

```python
# para preencher dados nulos por um valor predeterminado e deixa-los na base original
#fillna - preenche os valores nulos || inplace - mantem na base de dados original sem precisar aponta-la
vendas["estado"].fillna("MS", inplace=True)
vendas.info()
```

```python
# podemos ver que o preço continua com dados nulos, para este caso podemos substituir pela média
media_preco = vendas["Preço"].mean()

vendas["Preço"].fillna(media_preco, inplace=True)

vendas.info()
```

```python
#uma das premissas é que o preço nao pode ser maior que o preço com frete, para isso eliminamos
#todas as linhas que possuem esta caracteristica
vendas[["Preço","Preço_com_frete"]].describe()
```

```python
preco_errado = vendas.query("Preço > Preço_com_frete")

vendas_correto = vendas.query("Preço < Preço_com_frete")
```

```python
vendas_correto.describe()
```

```python
vendas_correto.head()
```

```python
# pode ser feita analise com filtros diretamente no pront utilizando a função .query
vendas_correto.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'").describe()
```

```python
# metricas a serem calculadas: Depatamentos mais vendidos:

vendas_correto.groupby("Nome_Departamento").idcompra.nunique().sort_values(ascending=False)
```

```python
# media de preço com frete por departamento

round(vendas_correto.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending=False),2)

# funçao round determina o numero de casas decimais do conjunto.
```

```python
# quantidade de vendas por mes:

vendas_correto.groupby("Data").idcompra.nunique().sort_values(ascending=False)
```

```python
# para agrupar por mes é necessario criar uma nova coluna extraindo o mes do dataset

vendas_correto["mes"] = vendas_correto["Data"].dt.month
vendas_correto.head()
```

```python
# agrupando pelo mes:

vendas_correto.groupby("mes").idcompra.nunique().sort_values(ascending=False)
```

```python
cliente = pd.read_excel("cliente_varejo.xlsx")
cliente.head()
```

```python
vendas_correto.head()
```

```python
# como vemos, as duas bases possuem a coluna 'cliente_Log' em comum

cliente[["idade","renda"]].describe()
```

```python
cliente.info()
```

```python
cliente = cliente.astype({"renda":"float"})
cliente.info()
```

```python
#join no python

vendas_cliente = vendas_correto.merge(cliente, how="left", on="cliente_Log")
vendas_cliente.head()
```

```python
# media de renda para canal de venda e media de idade de clientes por bandeira

agg_idcanal_renda = round(vendas_cliente.groupby("idcanalvenda")["renda"].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_idcanal_renda
```

```python
agg_idade_bandeira = round(vendas_cliente.groupby("bandeira")["idade"].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_idade_bandeira
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(7,4))

plt.bar(agg_idade_bandeira["bandeira"], agg_idade_bandeira["idade"], color ="green")

plt.ylabel("Média de Idade")

plt.title("Idade média por Bandeira", loc="center", fontsize=18, fontweight="bold", color="black");
```

```python
fig = plt.figure(figsize=(7,4))

plt.bar(agg_idcanal_renda["idcanalvenda"], agg_idcanal_renda["renda"], color="green")

plt.ylabel("Renda Média")

plt.title("Renda média por canal de venda", loc="center", fontsize=18, fontweight="bold",color="black");
```

```python
venda_por_data = vendas_correto.groupby("Data").idcompra.nunique().reset_index()
venda_por_data
```

```python
fig = plt.figure(figsize=(20,4))
plt.xlabel("Data da venda")
plt.ylabel("Quantidade de vendas")
x = venda_por_data["Data"]
y = venda_por_data["idcompra"]

plt.plot(x, y, color="green", linewidth=2)

plt.show()
```

```python
agg_dept_preco = round(vendas_correto.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending=False),2)
agg_dept_preco
```

```python
fig = plt.figure(figsize=(7,4))

plt.bar(agg_dept_preco["Nome_Departamento"], agg_dept_preco["Preço_com_frete"], color="green")

plt.ylabel("Preço médio")

plt.title("Renda média por departamento", loc="center", fontsize=18, fontweight="bold",color="black");
```
